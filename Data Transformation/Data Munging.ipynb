<B>What is data Munging?</B><BR>
Data munging(Cleanup) is the process of transofrmting and mapping raw data into usable(tidy) format with the intent of making it appropriate and valuable for
downstream purposes(Tranformation, Data Enrichment and outbound analytics/Reporting ) 



**Types of Data Munging:**<br>
a) **Active Data Munging** <br> 
   
b) **Passive Data Munging**<br>
   To perform analysis of RAW data to identify Patterns and Attributes<br>
   (Data Discovery/Data Exploration/EXploratory Data Analysis)
